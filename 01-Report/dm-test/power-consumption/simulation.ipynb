```python
import numpy as np
import matplotlib.pyplot as plt

T = 3600.0   # Simulation time (s)
r = 10.0     # Measurement rate (s)
P = 1      # Success probability factor
N_range = np.arange(0, 31) # Nbr of nearby devices (0 to 30)

MEAS_FACTOR = (T / r) * P

rssi_nbr_meas = [MEAS_FACTOR * n for n in N_range]
cs_nbr_meas = [MEAS_FACTOR * 1] * len(N_range)
print(rssi_nbr_meas)
print(cs_nbr_meas)
cs_nbr_meas[0] = 0

plt.figure(figsize=(10, 6))
plt.plot(N_range, rssi_nbr_meas, marker='o', linestyle='-', color='b', 
         label=f'RSSI-based measures')
plt.plot(N_range, cs_nbr_meas, marker='s', linestyle='--', color='r', 
         label=f'CS-based measures')

plt.title('Expected successful measures: RSSI vs CS Methods (1 hour simulation)', fontsize=14)
plt.xlabel('Number of nearby devices', fontsize=12)
plt.ylabel('Expected successful measures per device', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(np.arange(0, 31, 2))
plt.yticks(np.arange(0, 12000, 1000))
plt.legend(fontsize=10, loc='upper left')

plt.tight_layout()

plt.show()
```
